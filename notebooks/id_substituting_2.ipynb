```python
df_complete = pd.read_csv(r"..\data\csv\games.csv")
df_champs = pd.read_csv(r"..\data\csv\champ_file_2.csv")
```

```python
df_complete.columns
```

```python
df_complete[["winner", "firstBlood", "firstTower", "firstInhibitor", "firstBaron", "firstDragon", "firstRiftHerald"]] = df_complete[["winner", "firstBlood", "firstTower", "firstInhibitor", "firstBaron", "firstDragon", "firstRiftHerald"]].replace([1, 2, 0], ["blue", "red", "nobody"])
```

```python
#creating df with only match data
df_games = df_complete[["winner", "gameDuration", "firstBlood", "firstTower", "firstInhibitor", "firstBaron", "firstDragon", "firstRiftHerald", "t1_towerKills",  "t1_inhibitorKills", "t1_baronKills", "t1_dragonKills", "t1_riftHeraldKills", "t2_towerKills",  "t2_inhibitorKills", "t2_baronKills", "t2_dragonKills", "t2_riftHeraldKills"]]

#creating df with only champs data
df = df_complete[["winner","seasonId", "t1_champ1id", "t1_ban1", "t1_champ2id", "t1_ban2",  "t1_champ3id", "t1_ban3",  "t1_champ4id", "t1_ban4", "t1_champ5id", "t1_ban5", "t2_champ1id", "t2_ban1", "t2_champ2id","t2_ban2", "t2_champ3id", "t2_ban3", "t2_champ4id", "t2_ban4", "t2_champ5id", "t2_ban5"]]
```

```python
df.shape
```

```python
for i in range(51490):
    
    # champs picks
    df["t1_champ1id"][i] = df_champs[df_champs["id"] == df["t1_champ1id"][i]]["name"].values[0]
    df["t1_champ2id"][i] = df_champs[df_champs["id"] == df["t1_champ2id"][i]]["name"].values[0]
    df["t1_champ3id"][i] = df_champs[df_champs["id"] == df["t1_champ3id"][i]]["name"].values[0]
    df["t1_champ4id"][i] = df_champs[df_champs["id"] == df["t1_champ4id"][i]]["name"].values[0]
    df["t1_champ5id"][i] = df_champs[df_champs["id"] == df["t1_champ5id"][i]]["name"].values[0]
    df["t2_champ1id"][i] = df_champs[df_champs["id"] == df["t2_champ1id"][i]]["name"].values[0]
    df["t2_champ2id"][i] = df_champs[df_champs["id"] == df["t2_champ2id"][i]]["name"].values[0]
    df["t2_champ3id"][i] = df_champs[df_champs["id"] == df["t2_champ3id"][i]]["name"].values[0]
    df["t2_champ4id"][i] = df_champs[df_champs["id"] == df["t2_champ4id"][i]]["name"].values[0]
    df["t2_champ5id"][i] = df_champs[df_champs["id"] == df["t2_champ5id"][i]]["name"].values[0]

    # champs bans
    df["t1_ban1"][i] = df_champs[df_champs["id"] == df["t1_ban1"][i]]["name"].values[0]
    df["t1_ban2"][i] = df_champs[df_champs["id"] == df["t1_ban2"][i]]["name"].values[0]
    df["t1_ban3"][i] = df_champs[df_champs["id"] == df["t1_ban3"][i]]["name"].values[0]
    df["t1_ban4"][i] = df_champs[df_champs["id"] == df["t1_ban4"][i]]["name"].values[0]
    df["t1_ban5"][i] = df_champs[df_champs["id"] == df["t1_ban5"][i]]["name"].values[0]
    df["t2_ban1"][i] = df_champs[df_champs["id"] == df["t2_ban1"][i]]["name"].values[0]
    df["t2_ban2"][i] = df_champs[df_champs["id"] == df["t2_ban2"][i]]["name"].values[0]
    df["t2_ban3"][i] = df_champs[df_champs["id"] == df["t2_ban3"][i]]["name"].values[0]
    df["t2_ban4"][i] = df_champs[df_champs["id"] == df["t2_ban4"][i]]["name"].values[0]
    df["t2_ban5"][i] = df_champs[df_champs["id"] == df["t2_ban5"][i]]["name"].values[0]


```
